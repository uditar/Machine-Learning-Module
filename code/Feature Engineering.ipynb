```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../data/Zomato/zomato.csv",sep=',', encoding='latin-1')
pd.options.display.max_columns = None
columns=['Restaurant Name','Address','Locality','Locality Verbose','Rating color','Rating text','Votes','Switch to order menu']
df=df.drop(columns,axis=1)
df.fillna(method='bfill',inplace=True)
le = preprocessing.LabelEncoder()
df = df.apply(le.fit_transform)
df.dtypes
```

```python
corr_df = df.corr(method='spearman')
mask = np.zeros_like(corr_df)
mask[np.triu_indices_from(mask)] = True
# {‘pearson’, ‘kendall’, ‘spearman’}
print("--------------- CREATE A HEATMAP ---------------")
sns.set(font_scale=2.5)
plt.figure(figsize = (30,20))
with sns.axes_style("white"):
    sns_plot = sns.heatmap(corr_df,mask=mask,cmap='RdYlGn_r', vmax=1.0, vmin=-1.0,annot=True,annot_kws={"size": 20},square=True)
fig = sns_plot.get_figure()
fig.savefig("../features/HeatMapphase2.png")
```
